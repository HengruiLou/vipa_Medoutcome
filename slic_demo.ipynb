```python
import cv2
import matplotlib.pyplot as plt
```

```python
data_path = '/nfs3/lhm/demo.png'
img = cv2.imread(data_path)
plt.imshow(img)
plt.show()
```

```python
slic = cv2.ximgproc.createSuperpixelSLIC(img,region_size=30,ruler = 20.0)
slic.iterate(20)
mask_slic = slic.getLabelContourMask() #MaskMask==1
label_slic = slic.getLabels()        #
number_slic = slic.getNumberOfSuperpixels()  #
mask_inv_slic = cv2.bitwise_not(mask_slic)
img_slic = cv2.bitwise_and(img,img,mask =  mask_inv_slic) #
```

```python
plt.imshow(img_slic)
plt.show()
```

```python
plt.imshow(label_slic)
plt.show()
```

```python
seeds = cv2.ximgproc.createSuperpixelSEEDS(img.shape[1], img.shape[0], img.shape[2], 2000, 15, 3, 5, True)
seeds.iterate(img, 10)
mask_seeds= seeds.getLabelContourMask()
mask_inv_seeds = cv2.bitwise_not(mask_seeds)
img_seeds = cv2.bitwise_and(img,img,mask =  mask_inv_seeds)
```

```python
plt.imshow(img_seeds)
plt.show()
```
